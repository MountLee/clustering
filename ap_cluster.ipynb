# Compare my AP and several algo by sklearn
+ It's hard to say which one is better

```python
from sklearn.cluster import AffinityPropagation
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import Birch
```

## Case I: regular grid

```python
N = 10
x = []
for i in range(N):
    for j in range(N):
        x.append([i,j])
x = np.array(x)
N = x.shape[0]
x = x[np.random.permutation(N),:]
# x = np.random.rand(N,2)
S = np.zeros(shape = (N,N))
for i in range(N):
    for j in range(N):
        S[i,j] = -10 * sum((x[i,:] - x[j,:])**2)
p = np.median(S) * np.ones(N)
```

```python
result = ap_cluster(S,p)
visual_result(x,result[0])
```

```python
ap = AffinityPropagation(damping = 0.8, max_iter=1000, convergence_iter=100).fit(x)
visual_result(x,ap.labels_)
```

```python
kmeans = KMeans(n_clusters=10, n_init = 100, random_state=0).fit(x)
visual_result(x,kmeans.labels_)
```

```python
agglo = AgglomerativeClustering(n_clusters=10).fit(x)
visual_result(x,agglo.labels_)
```

```python
birch = Birch(threshold=0.5, branching_factor=50, n_clusters=10).fit(x)
visual_result(x,birch.labels_)
```
